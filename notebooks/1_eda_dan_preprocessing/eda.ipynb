```python
import pandas as pd

# Baca file Excel
df = pd.read_excel("Dataset Cerita Rakyat.xlsx")

# Hitung jumlah kemunculan setiap daerah
asal_counts = df['asal'].value_counts()

# Tampilkan hasilnya
print(asal_counts)

# (Opsional) Tampilkan sebagai bar chart
import matplotlib.pyplot as plt

asal_counts.plot(kind='bar', title='Persebaran Asal Cerita Rakyat', xlabel='Daerah', ylabel='Jumlah')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load file
df = pd.read_csv("cerita_rakyat_tokenized_clean.csv")  # Ganti dengan nama file aslimu

# Hitung jumlah kalimat unik per cerita
jumlah_per_story = df.groupby('story_id')['sentence_id'].nunique().reset_index(name='jumlah_kalimat')

# Hitung total seluruh kalimat dari semua cerita
total_kalimat = jumlah_per_story['jumlah_kalimat'].sum()

# Tampilkan hasil
print(jumlah_per_story)
print("\nTotal seluruh kalimat dalam dataset:", total_kalimat)
```

```python
import pandas as pd

# Load file tokenized
df = pd.read_csv("cerita_rakyat_tokenized_clean.csv")  # Ganti nama file sesuai milikmu

# Hitung jumlah kata per cerita
kata_per_cerita = df.groupby('story_id').size()

# Hitung median dari jumlah kata per cerita
median_kata = kata_per_cerita.median()

print("Median kata per cerita:", median_kata)
```

```python
import pandas as pd

# Load file tokenized
df = pd.read_csv("cerita_rakyat_tokenized_clean.csv")  # ganti nama sesuai file kamu

# Hitung jumlah kata per kalimat
kata_per_kalimat = df.groupby(['story_id', 'sentence_id']).size()

# Hitung median
median_kata_per_kalimat = kata_per_kalimat.median()

print("Median kata per kalimat:", median_kata_per_kalimat)
```
